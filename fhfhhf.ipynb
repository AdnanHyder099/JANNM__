```python
# Question 9
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    if x<=-1:
        return(20)
    elif -1<=x<=1:
        return( -5*x + 10 )
    elif 1<=x<=3:
        return( -10*x**2 + 35*x - 20 )
    elif 3<=x<=4:
        return( -5*x + 10 )
    elif x>=4:
        return(-10)
```

```python
x = [-1,0,1,2,3,4]
y = []
for i in x:
    y.append(f(i))
plt.plot(x,y)
plt.show()
```
